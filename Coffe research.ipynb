## Task 3 Каковы корреляции между различными показателями оценки кофе?

```python
x = np.array(res['aroma'])
x1 = np.array(res['aroma'])
x2 = np.array(res['flavor'])
x3 = np.array(res['acidity'])
x4 = np.array(res['aroma'])
x5 = np.array(res['flavor'])
x6 = np.array(res['acidity'])
x7 = np.array(res['moisture'])
```

```python
y = np.array(res['flavor'])
y1 = np.array(res['aftertaste'])
y2 = np.array(res['aftertaste'])
y3 = np.array(res['aftertaste'])
y4 = np.array(res['sweetness'])
y5 = np.array(res['sweetness'])
y6 = np.array(res['body'])
y7 = np.array(res['clean_cup'])
```

```python
f, ax = plt.subplots(2,4, figsize=(25, 10))
ax[0,0].scatter(x,y, s=10, c='g')
ax[0,0].set_title('Aroma - Flavor')
ax[0,0].grid()
ax[0,0].set(xlabel='Aroma', ylabel='Flavor')
ax[0,0].set_xlim([6, 9])
ax[0,0].set_ylim([6, 9])

ax[0,1].scatter(x1,y1, s=10, c='b')
ax[0,1].set_title('Aroma - Aftertaste')
ax[0,1].grid()
ax[0,1].set(xlabel='Aroma', ylabel='Aftertaste')
ax[0,1].set_xlim([6, 9])
ax[0,1].set_ylim([6, 9])

ax[0,2].scatter(x2,y2, s=10, c='r')
ax[0,2].set_title('Flavor - Aftertaste')
ax[0,2].grid()
ax[0,2].set(xlabel='Flavor', ylabel='Aftertaste')
ax[0,2].set_xlim([6, 9])
ax[0,2].set_ylim([6, 9])

ax[0,3].scatter(x3,y3, s=10, c='m')
ax[0,3].set_title('Acidity - Aftertaste')
ax[0,3].grid()
ax[0,3].set(xlabel='Acidity', ylabel='Aftertaste')
ax[0,3].set_xlim([6, 9])
ax[0,3].set_ylim([6, 9])

ax[1,0].scatter(x4,y4, s=10, c='c')
ax[1,0].set_title('Aroma - Sweetness')
ax[1,0].grid()
ax[1,0].set(xlabel='Aroma', ylabel='Sweetness')
ax[1,0].set_xlim([6, 9])
ax[1,0].set_ylim([6, 11])

ax[1,1].scatter(x5,y5, s=10, c='k')
ax[1,1].set_title('Flavor - Sweetness')
ax[1,1].grid()
ax[1,1].set(xlabel='Flavor', ylabel='Sweetness')
ax[1,1].set_xlim([6, 9])
ax[1,1].set_ylim([6, 11])

ax[1,2].scatter(x6,y6, s=10, c='m')
ax[1,2].set_title('Acidity - Body')
ax[1,2].grid()
ax[1,2].set(xlabel='Acidity', ylabel='Body')
ax[1,2].set_xlim([6, 9])
ax[1,2].set_ylim([6, 9])

ax[1,3].scatter(x7,y7, s=10, c='b')
ax[1,3].set_title('Moisture - Clean cup')
ax[1,3].grid()
ax[1,3].set(xlabel='Moisture', ylabel='Clean cup')

```

### Выводы: в данном задании мы проанализировали взаимосвязь некоторых показателей оценки кофе. Из графиков выше можем сказать, что:
  - Взаимосвязи между показателями Aroma - Flavor, Aroma - Aftertaste, Flavor - Aftertaste, Acidity - Aftertaste и Acidity - Body имеют положительный коэфициент корреляции (положительную взаимосвязь).
  - А взаимосвязь между показателями Aroma - Sweetness, Flavor - Sweetness и Moisture - Clean cup отсутствует.

# Task 6 Существенно ли влияет высота на качество кофе?

```python
res6 = res[['total_cup_points', 'altitude_mean_meters']]
```

```python
cnt6 = res6.groupby(['altitude_mean_meters'])['total_cup_points'].mean() #Высчитываем среднюю оценку кофе по каждой высоте (средней высоте)
```

```python
x = np.array(cnt6.index)
y = np.array(cnt6)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y, s=10, c='b')
ax.grid()
ax.set(xlabel='Средняя высота', ylabel='Средняя оценка кофе')
ax.set_xlim([0, 6000])
```

### Вывод: как видно из диаграммы выше, влияние высоты на оценку (качество) кофе отсутствует (коэфициент корреляции стримится к 0)
